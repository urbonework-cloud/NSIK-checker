```python
from google.colab import files  # ✅ Import the files module

print("📁 Please upload your IFC file...")
uploaded = files.upload()  # Opens file picker in Colab

# Get the uploaded file name
ifc_path = list(uploaded.keys())[0]
print(f"✅ Loaded file: {ifc_path}")
```

```python
import ifcopenshell
# Load the IFC file
ifc = ifcopenshell.open(ifc_path)

```

```python
TARGET_PSET = "PG"

# Required parameters in PG
required_params = [
    "A.01_Pavadinimas",
    "A.02_Aukštas",
    "A.03_Patalpos numeris",
    "A.05_TS. Nr.",
    "A.06_Elemento tipas",
    "B.01_Aukštis",
    "B.02_Plotis",
    "B.03_Ilgis",
    "B.05_Diametras",
    "B.07_Plotas",
    "B.08_Tūris",
    "C.01_Medžiagiškumas",
    "D.01_Funkcinės sistemos tipas",
    "D.02_Techninės sistemos tipas",
    "D.03_Elemento tipo kodas",
    "D.04_Patalpos kodas",
    "E.01_Specifiniai rekalavimai",
    "E.04_Galingumas",
    "E.08_Akustiniai reikalavimai",
    "F.02_Apdailos A plotas",
    "F.03_Apdaila B",
    "F.04_Apdailos B plotas",
    "G.01_Atsparumas ugniai"
]

def get_psets(element):
    """Return all property sets attached to an IFC element."""
    psets = {}
    for rel in getattr(element, "IsDefinedBy", []):
        if rel.is_a("IfcRelDefinesByProperties"):
            prop = rel.RelatingPropertyDefinition
            if prop.is_a("IfcPropertySet"):
                psets[prop.Name] = prop
    return psets


print(f"Checking required parameters in PSet '{TARGET_PSET}'...\n")

for element in ifc.by_type("IfcObject"):
    name = element.Name if element.Name else "(no name)"
    psets = get_psets(element)

    if TARGET_PSET not in psets:
        continue  # skip elements without PG

    pset = psets[TARGET_PSET]

    # Collect existing parameter names
    existing_params = {prop.Name for prop in pset.HasProperties}

    # Find missing parameters
    missing = [p for p in required_params if p not in existing_params]

    print(f"=== {element.is_a()} | Name: {name} ===")

    if missing:
        print("❌ Missing parameters:")
        for m in missing:
            print("   -", m)
    else:
        print("✅ All required parameters are present")

    print()
```
